**Important:** These notebooks are saved outside of the pd, and will disappear once your workspace session is over. **If you'd like to save changes to these notebooks, be sure to move them to your `pd` folder before terminating the workspace.**

```python
import ipywidgets as widgets
from IPython.display import display
import subprocess
import os
import shutil
import datetime

clone_button = widgets.Button(description="Get Notebooks", button_style='success')
output = widgets.Output()

def on_button_clicked(b):
    with output:
        output.clear_output()
        url = 'https://github.com/uc-cdis/heal-notebooks.git'
        target_dir = './HEAL-notebooks'
        subdir = 'notebooks'
        
        # Create a backup if there's already an existing directory
        if os.path.exists(target_dir):
            # Create a backup dir name with timestamp
            timestamp = datetime.datetime.now().strftime('%Y%m%d_%H%M%S')
            backup_dir = f"{target_dir}_backup_{timestamp}"
            # Copy the contents of the existing directory to backup dir
            shutil.move(target_dir, backup_dir)
            print(f"Backup of existing HEAL-notebooks directory created at: {backup_dir}")
        
        print(f'Cloning directory "{subdir}" from {url} into {target_dir} ...')
        try:
            # Use sparse-checkout to populate the target directory with only the desired dir, not the whole repo
            # prepare to do a sparse checkout - initialize empty repo with blobless filter
            subprocess.run(['git', 'clone',
                            '--filter=blob:none',  # don't download any file
                            '--no-checkout',  # don't checkout any files
                            url, target_dir],
                           check=True, capture_output=True, text=True)
            
            # Configure sparse-checkout patterns
            # turn off the default cone mode to do strict pattern matching only
            # Use 'notebooks/' to ensure git treats the checkput pattern as explicitly a directory
            subprocess.run(['git', '-C', target_dir,
                            'sparse-checkout', 'set', '--no-cone', f"{subdir}/"],
                           check=True, capture_output=True, text=True)
            
            # Checkout the files
            subprocess.run(['git', '-C', target_dir, 'checkout'], 
                           check=True, capture_output=True, text=True)
            
            print(f'Directory "{subdir}" cloned successfully into {target_dir}.')
            
        except subprocess.CalledProcessError as e:
            # check=True in earlier subprocess.run lines routes any non-zero exit codes here
            print(f"Git Command Failed: {e.stderr.strip()}")
        except Exception as e:
            print(f'An unexpected error occurred: {e}')

clone_button.on_click(on_button_clicked)
display(widgets.VBox([clone_button, output]))
```

### Download files from the HEAL Discovery page to the Workspace  

[Learn how to download data files through the HEAL Discovery page](https://heal.github.io/platform-documentation/downloading_files/) in our documentation. 

When you download files from the [HEAL Discovery Page](https://healdata.org/portal) to the workspace, these will be placed in the folder `healdata.org` in the `data` folder. These will not persist beyond the current workspace session if they stay here. Move these files to the `pd` directory if you want to access them again after you terminate your workspace session.

### Install software tools for analysis in the workspace  

Install software tools by using `pip install` (Python) or `CRAN` (R). If you have a requirements text, you can upload your requirements text and conveniently install all of them with `pip install -r requirements.txt`. 

The HEAL Workspace comes with some pre-installed software packages. You can view all pre-installed tools by opening a terminal window and using the command `pip list`.

### Upload files from local storage    

You can [upload files to the workspace](https://heal.github.io/platform-documentation/workspaces/heal_workspaces/#upload-save-and-download-filesnotebooks) by clicking the upload arrow icon in the upper left corner

If the locally-stored files you wish to analyze are large in number and/or size, you may need to zip them before uploading to a workspace. Once in a workspace, files can be unzipped using the python library [zipfile](https://docs.python.org/3/library/zipfile.html).
