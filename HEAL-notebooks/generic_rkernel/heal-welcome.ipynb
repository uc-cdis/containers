## Adding HEAL example notebooks to your `pd` folder

The [HEAL Example Analyses](https://uc-cdis.github.io/heal-notebooks-pages/) notebooks are stored in [GitHub](https://github.com/uc-cdis/heal-notebooks-pages), and may be cloned into your `pd` folder so that you may try them out and experiment with them. To do so, use the Run &rarr; Run All Cells menu item, and then click on the green Get Notebooks button that appears below.

```python
import ipywidgets as widgets
from IPython.display import display
import subprocess

clone_button = widgets.Button(description="Get Notebooks", button_style='success')
output = widgets.Output()

def on_button_clicked(b):
    with output:
        output.clear_output()
        url = 'https://github.com/uc-cdis/heal-notebooks-pages.git'
        print(f'Cloning {url}...')
        try:
            result = subprocess.run(['git', 'clone', '--recurse-submodules', '-j8', url, './pd/HEAL-notebooks'],
                                    capture_output=True, text=True)
            if result.returncode == 0:
                print('Repository cloned successfully.')
            else:
                print(f'Error: {result.stderr}')
        except Exception as e:
            print(f'An error occurred: {e}')

clone_button.on_click(on_button_clicked)
display(widgets.VBox([clone_button, output]))
```
